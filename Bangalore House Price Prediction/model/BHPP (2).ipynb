### Explore total_sqft feature

```python
def is_float(x):
    """
    Check if the input can be converted to a floating-point number.

    Args:
        x (str or numeric): The input value to be checked.

    Returns:
        bool: True if the input can be converted to a float, False otherwise.
    """
    try:
        float(x)
    except ValueError:  # Catch the ValueError specifically when conversion fails
        return False
    return True
```

```python
# We want to filter out rows where the 'total_sqft' column does not represent a floating-point number
# Apply the 'is_float' function to each value in the 'total_sqft' column
# and use the negation operator (~) to get rows where the function returns False
filtered_rows = df1[~df1['total_sqft'].apply(is_float)]

# Display the first 10 rows of the filtered DataFrame
print(filtered_rows.head(10))
```

```python
def convert_sqft_to_num(x):
    # Split the input string by the hyphen '-' character
    tokens = x.split('-')

    # Check if the input contains a range (two values separated by a hyphen)
    if len(tokens) == 2:
        # Calculate the average of the two values and return it
        return (float(tokens[0]) + float(tokens[1])) / 2

    try:
        # Try to convert the input string to a float (for single values)
        return float(x)
    except:
        # If the input cannot be converted to a float, return None
        return None

# Example usage
input_value = "2100-2850"
result = convert_sqft_to_num(input_value)
print(result)  # This will print the average of 2210 and 2850: 2475.0
```

```python
# Make a copy of the DataFrame df1 and store it in df2
df2 = df1.copy()

# Apply the convert_sqft_to_num function to each value in the 'total_sqft' column of df1
df2['total_sqft'] = df1['total_sqft'].apply(convert_sqft_to_num)

# Display rows 29 to 30 of the modified DataFrame df2
print(df2[29:31])
```

**Outlier Removal Using Standard Deviation and Mean**

```python
# Define a function to remove outliers from a DataFrame based on the price per square foot (pps) column
def remove_pps_outliers(df):
    # Create an empty DataFrame to store the filtered data
    df_out = pd.DataFrame()

    # Iterate through each unique location in the DataFrame
    for key, subdf in df.groupby('location'):
        # Calculate the mean and standard deviation of price per square foot for the current location
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)

        # Filter out rows where price per square foot is beyond one standard deviation from the mean
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]

        # Concatenate the filtered data for the current location to the output DataFrame
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)

    # Return the DataFrame with outliers removed
    return df_out

# Call the remove_pps_outliers function on the DataFrame df4 and store the result in df5
df5 = remove_pps_outliers(df4)

# Print the shape of the filtered DataFrame df5
print(df5.shape)
```

Now we can remove those 2 BHK apartments whose price_per_sqft is less than mean price_per_sqft of 1 BHK apartment

```python
def remove_bhk_outliers(df):
    # Initialize an array to store indices of data points to exclude
    exclude_indices = np.array([])

    # Group the DataFrame by the 'location' column
    for location, location_df in df.groupby('location'):
        # Create a dictionary to store statistics for each BHK configuration
        bhk_stats = {}

        # Iterate over the grouped data for each BHK configuration
        for bhk, bhk_df in location_df.groupby('bhk'):
            # Calculate mean, standard deviation, and count for price per sqft
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        # Iterate again over the grouped data for each BHK configuration
        for bhk, bhk_df in location_df.groupby('bhk'):
            # Get statistics for the previous BHK configuration (bhk-1)
            stats = bhk_stats.get(bhk-1)

            # Check if stats are available and if the count is greater than 5
            if stats and stats['count'] > 5:
                # Find indices of data points with price_per_sqft lower than the mean of the previous BHK configuration
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < stats['mean']].index.values)

    # Drop the data points with outlier prices using the exclude_indices array
    return df.drop(exclude_indices, axis='index')

# Call the function and store the result in df6
df6 = remove_bhk_outliers(df5)

# Print the shape of the resulting DataFrame
print(df6.shape)
```

#**Build a Model Now...**

```python
X = df10.drop(['price'],axis='columns')
X.head(3)
```

```python
y = df10.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X.values,y.values,test_size=.20,random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score



# Ridge Regression
ridge_model = Ridge(alpha=1.0)  # You can adjust the alpha (regularization strength)
ridge_model.fit(X_train, y_train)
ridge_predictions = ridge_model.predict(X_test)
ridge_score = r2_score(y_test, ridge_predictions)
print("Ridge Regression R2 Score:", ridge_score)

# Lasso Regression
lasso_model = Lasso(alpha=1.0)  # You can adjust the alpha (regularization strength)
lasso_model.fit(X_train, y_train)
lasso_predictions = lasso_model.predict(X_test)
lasso_score = r2_score(y_test, lasso_predictions)
print("Lasso Regression R2 Score:", lasso_score)

# Elastic Net Regression
elastic_net_model = ElasticNet(alpha=1.0, l1_ratio=0.5)  # You can adjust alpha and l1_ratio
elastic_net_model.fit(X_train, y_train)
elastic_net_predictions = elastic_net_model.predict(X_test)
elastic_net_score = r2_score(y_test, elastic_net_predictions)
print("Elastic Net Regression R2 Score:", elastic_net_score)

# Decision Tree Regression
dt_model = DecisionTreeRegressor(max_depth=None)  # You can adjust max_depth and other parameters
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_score = r2_score(y_test, dt_predictions)
print("Decision Tree Regression R2 Score:", dt_score)

# Random Forest Regression
rf_model = RandomForestRegressor(n_estimators=100, max_depth=None)  # You can adjust n_estimators, max_depth, etc.
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_score = r2_score(y_test, rf_predictions)
print("Random Forest Regression R2 Score:", rf_score)
```

###**Use K Fold cross validation to measure accuracy of our LinearRegression model**

```python
# Import the necessary libraries
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

# Create a ShuffleSplit cross-validation object with 5 splits
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

# Perform cross-validation using the linear regression model
# X is the feature matrix and y is the target vector
# The 'cv' parameter specifies the cross-validation strategy
# 'cross_val_score' returns an array of scores for each fold
cross_val_scores = cross_val_score(LinearRegression(), X, y, cv=cv)

# Print the cross-validation scores
print("Cross-validation scores:", cross_val_scores)
```

We can see that in 5 iterations we get a score above 80% all the time.

**Test the model for few properties**

```python
def price_predict(location, sqft, bath, BHK):
    # Find the index of the 'location' column in the feature matrix
    loc_index = np.where(X.columns == location)[0][0]

    # Create a new feature vector filled with zeros
    x = np.zeros(len(X.columns))

    # Fill in the feature vector with input values
    x[0] = sqft
    x[1] = bath
    x[2] = BHK

    # If the 'location' exists in the feature matrix, set its value to 1
    if loc_index >= 0:
        x[loc_index] = 1

    # Predict the price using the trained linear regression model 'lr_clf'
    predicted_price = lr_clf.predict([x])[0]

    return predicted_price
```
